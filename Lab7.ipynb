```python
url = 'http://www.gutenberg.org/files/3176/3176-0.txt'
response = request.urlopen(url)
raw = response.read().decode('utf8')
print(f'{type(raw)}, \n{len(raw)}, \n{raw[:501]}')
```

```python
places = GeoText(raw)
cities = list(places.cities)
cities #Calling this shows my list of cities. You could also use print(cities)
```

```python
geolocator = Nominatim(user_agent='parashararpan25@gmail.com', timeout=2) # requires your e-mail address to track usage. Timeout set for 2 seconds.

lat_lon = []
for city in cities:
  try:
    location = geolocator.geocode(city)
    if location:
      # If you want to see only lat/lon as they process you can include the following code here but I found it added a lot of time to the operation: print(location.latitude, location.longitude)
      lat_lon.append(location)
  except GeocoderTimedOut as e:
    print("Error: geocode failed on input %s with message %s">
          (city, e))
lat_lon
```

```python
df = pd.DataFrame(lat_lon, columns=['City Name', 'Coordinates'])
df.head(7)
```

```python
geometry = [Point(x[1], x[0]) for x in df['Coordinates']]
geometry[:7]
```

```python
# coordinate system we're using
crs = {'init': 'epsg:4326'}

# convert df to geo df
geo_df = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
geo_df.head()
```

```python
import geopandas as gpd
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from urllib import request
from geotext import GeoText

from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

from shapely.geometry import Point, Polygon
import descartes

import nltk
import seaborn as sns

```

```python
def mapcity(url):
 
  response = request.urlopen(url)                            # Makes text request from url
  raw = response.read().decode('utf8')
 
  places = GeoText(raw)                                      # Extracts city names from text
  citylist = list(places.cities)
 
  geolocator = Nominatim(user_agent='parashararpan25@gmail.com', timeout=2)    #Gets latitudes and longitudes
  lat_lon = []
  for city in citylist:
    try:
      location = geolocator.geocode(city)
      if location:
        lat_lon.append(location)
    except GeocoderTimedOut as e:
      print("Error: geocode failed on input %s with message %s">
            (city, e))
  
  
  %matplotlib inline                 
  sns.set()

  
  freqdist1 = nltk.FreqDist(citylist)      # Create freq dist and plot
  freqdist1.plot(20) #running w/out argument plots all words! Specifying the top 20
  
 
  df = pd.DataFrame(lat_lon, columns=['City Name', 'Coordinates'])    # Creates data frame of data

  geometry = [Point(x[1], x[0]) for x in df['Coordinates']]

  crs = {'init': 'epsg:4326'}           # Setting the coordinate system
  
  geo_df = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)  # convert df to geo df

  # make the map
  countries_map = gpd.read_file('Countries_WGS84.shp')
  
  f, ax = plt.subplots(figsize=(16, 16))
  countries_map.plot(ax=ax, alpha=0.3, color='grey')
  geo_df['geometry'].plot(ax=ax, markersize = 30, color = 'r', marker = '^', alpha=.4)
```

```python
mapcity('https://www.gutenberg.org/cache/epub/103/pg103.txt')
```
